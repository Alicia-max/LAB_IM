```python
def boot_select(LC_index, boot_res, X, level=2.5): 
    """
   
    Select the important voxels based on the boot stability scores
    
    Inputs
    -------
    LC_index : int
        integer indices of the latent variables
    boot_res : pandas DataFrame
        Dataframe with the bootstrap stability scores
    X : pandas DataFrame
        Dataframe with the original input data
    level : float
        The cutoff threshold for selecting voxels
    
    Ouput
    -------
    select_X : numpy array
        Array of selected voxels
    """
    select_X=np.zeros(len(X.iloc[:, LC_index]))
  
    index = np.argwhere(np.array(abs(boot_res.iloc[:,LC_index]))>level)
    print(index)
    select_X[index]=X.iloc[index[:,0], LC_index]
    print(select_X)
    return np.array(select_X)
```

```python
def plot_behav(LC_indexes, x, U, name, err): 
    """
    Plot and Save the Behavioral Saliences
    
    Inputs
    -------
    LC_indexes : list
        List of tuples with the indexes of the LVs to plot
    x : array-like
        Array of labels for the behavioral variables
    U : array-like
        Matrix of behavioral saliencies
    name : str
        Name of the saliency map
    """
    for i in LC_indexes :
        f, ax = plt.subplots(figsize=(6,3))
        print(err[i[0]])
        fig=plt.bar(x, U[i[0]],yerr=err[i[0]],  color='lightseagreen')
        ax.set_ylabel(f"Loadings {i[0]+1}")
        plt.xticks(rotation=80)
        plt.savefig(f"../Plots/Behav/Behav{i[0]+1}_{name}.png", bbox_inches='tight')
        
        
```

```python
def boot_rel(orig, u_sum, u_square, n_boot):
    """
    Determines bootstrap ratios (BSR) of saliences from bootstrap distributions
    Parameters
    ----------
    orig : (G, L) array_like
        Original singular vectors
    u_sum : (G, L) array_like
        Sum of bootstrapped singular vectors
    u_square : (G, L) array_like
        Sum of squared bootstraped singular vectors
    n_boot : int
        Number of bootstraps used in generating `u_sum` and `u_square`
    Returns
    -------
    bsr : (G, L) `numpy.ndarray`
        Bootstrap ratios for provided singular vectors
    """

    u_sum2 = (u_sum ** 2) / n_boot
    u_se = np.sqrt(np.abs(u_square - u_sum2) / (n_boot - 1))
    bsr = orig / u_se

    return bsr, u_se
```
